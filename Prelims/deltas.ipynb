```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('reorganised_data.csv',index_col=0)
```

```python
df.head()
```

```python
df.columns
```

```python
names = ['BMI', 'weight', 'UCS', 'ECS', 'RCS']
changes = ['BMI1', 'BMI2', 'BMI3',
       'weight1', 'weight2', 'weight3', 'UCS1', 'UCS2', 'UCS3', 'ECS1', 'ECS2',
       'ECS3', 'RCS1', 'RCS2', 'RCS3']
df['delBMI1'] = df['BMI1'] - df['BMI']
df['delBMI2'] = df['BMI2'] - df['BMI']
df['delBMI3'] = df['BMI3'] - df['BMI']

df['delweight1'] = df['weight1'] - df['weight']
df['delweight2'] = df['weight2'] - df['weight']
df['delweight3'] = df['weight3'] - df['weight']

df['delUCS1'] = df['UCS1'] - df['UCS']
df['delUCS2'] = df['UCS2'] - df['UCS']
df['delUCS3'] = df['UCS3'] - df['UCS']

df['delECS1'] = df['ECS1'] - df['ECS']
df['delECS2'] = df['ECS2'] - df['ECS']
df['delECS3'] = df['ECS3'] - df['ECS']

df['delRCS1'] = df['RCS1'] - df['RCS']
df['delRCS2'] = df['RCS2'] - df['RCS']
df['delRCS3'] = df['RCS3'] - df['RCS']
```
