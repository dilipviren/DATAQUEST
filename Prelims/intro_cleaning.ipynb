```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('CaseStudyStatP23.xlsx')
df.head()
```

```python
datalist = []
for i in df['ID'].unique():
    x = df[df['ID']==i]
    x.reset_index(inplace=True)
    
    x['BMI1'] = x['BMI'][1]
    x['BMI2'] = x['BMI'][2]
    x['BMI3'] = x['BMI'][3]
    
    x['weight1'] = x['weight'][1]
    x['weight2'] = x['weight'][2]
    x['weight3'] = x['weight'][3]
    
    x['UCS1'] = x['UCS'][1]
    x['UCS2'] = x['UCS'][2]
    x['UCS3'] = x['UCS'][3]
    
    x['ECS1'] = x['ECS'][1]
    x['ECS2'] = x['ECS'][2]
    x['ECS3'] = x['ECS'][3]
    
    x['RCS1'] = x['RCS'][1]
    x['RCS2'] = x['RCS'][2]
    x['RCS3'] = x['RCS'][3]
    
    rowlist = x.iloc[0].tolist()
    datalist.append(rowlist)
```

```python
colnames = ['Index', 'ID', 'Treatment', 'Gender', 'Visit', 
             'Age', 'BMI', 'weight', 'Race',
             'Country', 'UCS', 'ECS', 'RCS', 'MedIncome',
             'BMI1','BMI2','BMI3','weight1','weight2','weight3',
             'UCS1','UCS2','UCS3','ECS1','ECS2','ECS3',
             'RCS1','RCS2','RCS3']
len(datalist[1]) == len(colnames)
```

```python
data = pd.DataFrame(datalist, columns=colnames)
data.head()
```

```python
data.drop('Index',axis=1,inplace=True)
```
